```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('householdtask3.csv')
```

```python
data.head(10)
```

```python
# Line chart

plt.plot(data['income'])
plt.plot(data['expenditure'])

# Title
plt.title("LINE CHART")

# x label and y label 
plt.xlabel('income')
plt.ylabel('expenditure')

#Legend
plt.legend(["income","expenditure"])


# Result
plt.show()
```

```python
# Bar chart

plt.bar(data['own_prop'], data['own_wm_prop'])

# Title
plt.title("BAR CHART")

# x label and y label
plt.xlabel('own_prop')               
plt.ylabel('own_wm_prop')

# Result
plt.show()
```
